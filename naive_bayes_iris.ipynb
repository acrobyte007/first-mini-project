```python
def gausian(df,f1,f2,f3,f4):
  df_setosa=df[df['class']=='iris-setosa']
  df_versicolor=df[df['class']=='iris-versicolor']
  df_virginica=df[df['class']=='iris-virginica']
  ## MEAN OF SETOSA
  f1_mean_setosa=df_setosa['sepal length'].mean()
  f2_mean_setosa=df_setosa['sepal width'].mean()
  f3_mean_setosa=df_setosa['petal length'].mean()
  f4_mean_setosa=df_setosa['petal width'].mean()
  ## STD OF SETOSA
  f1_std_setosa=df_setosa['sepal length'].std()
  f2_std_setosa=df_setosa['sepal width'].std()
  f3_std_setosa=df_setosa['petal length'].std()
  f4_std_setosa=df_setosa['petal width'].std()

  ## MEAN OF versicolor
  f1_mean_versicolor=df_versicolor['sepal length'].mean()
  f2_mean_versicolor=df_versicolor['sepal width'].mean()
  f3_mean_versicolor=df_versicolor['petal length'].mean()
  f4_mean_versicolor=df_versicolor['petal width'].mean()
  ## STD OF versicolor
  f1_std_versicolor=df_versicolor['sepal length'].std()
  f2_std_versicolor=df_versicolor['sepal width'].std()
  f3_std_versicolor=df_versicolor['petal length'].std()
  f4_std_versicolor=df_versicolor['petal width'].std()

  ## MEAN OF virginica
  f1_mean_virginica=df_virginica['sepal length'].mean()
  f2_mean_virginica=df_virginica['sepal width'].mean()
  f3_mean_virginica=df_virginica['petal length'].mean()
  f4_mean_virginica=df_virginica['petal width'].mean()
  ## STD OF virginica
  f1_std_virginica=df_virginica['sepal length'].std()
  f2_std_virginica=df_virginica['sepal width'].std()
  f3_std_virginica=df_virginica['petal length'].std()
  f4_std_virginica=df_virginica['petal width'].std()

  prob_setosa=len(df_setosa)/len(df)
  prob_versicolor=len(df_versicolor)/len(df)
  prob_virginica=len(df_virginica)/len(df)

  prob_setosa_f1=(1/(f1_std_setosa*np.sqrt(2*3.14))*np.exp(-(f1-f1_mean_setosa)**2)/2*(f1_std_setosa)**2)
  prob_setosa_f2=(1/(f2_std_setosa*np.sqrt(2*3.14))*np.exp(-(f2-f2_mean_setosa)**2)/2*(f2_std_setosa)**2)
  prob_setosa_f3=(1/(f3_std_setosa*np.sqrt(2*3.14))*np.exp(-(f3-f3_mean_setosa)**2)/2*(f3_std_setosa)**2)
  prob_setosa_f4=(1/(f4_std_setosa*np.sqrt(2*3.14))*np.exp(-(f4-f2_mean_setosa)**2)/2*(f4_std_setosa)**2)

  prob_versicolor_f1=(1/(f1_std_versicolor*np.sqrt(2*3.14))*np.exp(-(f1-f1_mean_versicolor)**2)/2*(f1_std_versicolor)**2)
  prob_versicolor_f2=(1/(f2_std_versicolor*np.sqrt(2*3.14))*np.exp(-(f2-f2_mean_versicolor)**2)/2*(f2_std_versicolor)**2)
  prob_versicolor_f3=(1/(f3_std_versicolor*np.sqrt(2*3.14))*np.exp(-(f3-f3_mean_versicolor)**2)/2*(f3_std_versicolor)**2)
  prob_versicolor_f4=(1/(f4_std_versicolor*np.sqrt(2*3.14))*np.exp(-(f4-f4_mean_versicolor)**2)/2*(f4_std_versicolor)**2)

  prob_virginica_f1=(1/(f1_std_virginica*np.sqrt(2*3.14))*np.exp(-(f1-f1_mean_virginica)**2)/2*(f1_std_virginica)**2)
  prob_virginica_f2=(1/(f2_std_virginica*np.sqrt(2*3.14))*np.exp(-(f2-f2_mean_virginica)**2)/2*(f2_std_virginica)**2)
  prob_virginica_f3=(1/(f3_std_virginica*np.sqrt(2*3.14))*np.exp(-(f3-f3_mean_virginica)**2)/2*(f3_std_virginica)**2)
  prob_virginica_f4=(1/(f4_std_virginica*np.sqrt(2*3.14))*np.exp(-(f4-f4_mean_virginica)**2)/2*(f4_std_virginica)**2)

  total_prob_setosa=prob_setosa*prob_setosa_f1*prob_setosa_f2*prob_setosa_f3*prob_setosa_f4
  total_prob_versicolor=prob_versicolor*prob_versicolor_f1*prob_versicolor_f2*prob_versicolor_f3*prob_versicolor_f4
  total_prob_virginica=prob_virginica*prob_virginica_f1*prob_virginica_f2*prob_virginica_f3*prob_virginica_f4

  if total_prob_setosa>total_prob_versicolor and total_prob_setosa>total_prob_virginica:
    return 'It is Iris_setosa'

  elif total_prob_versicolor>total_prob_setosa and total_prob_versicolor>total_prob_virginica:
    return 'It is Iris_versicolor'
  else :
    return 'It is Iris_virginica'
```
